### Data Cleaning

```python
apps_df.isna().sum()             # check on null values
```

Here, we realized that there are 1474 rows having null values under column 'Rating'. Hence, we replace the null values with median of overall 'Rating' values.

```python
# checking how many rating are more than 5 outlier
apps_df[apps_df["Rating"]>5]
```

```python
apps_df[10470:10475]
```

```python
median_ratings=apps_df["Rating"].median()
median_ratings
```

```python
apps_df["Rating"].fillna(median_ratings,inplace=True)
```

```python
apps_df.isnull().sum()
```

```python
# remove the record where 'Type' is having null value
index = apps_df[apps_df['Type'].isna()].index
apps_df.drop(axis=0, inplace=True, index=index)
```

```python
# checking on the null values
apps_df.isna().sum()
```

##### Mode of Categorical Columns

```python
#print(apps_df['Type'].mode())
print(apps_df['Current Ver'].mode())
print(apps_df['Android Ver'].mode())
```

```python
#apps_df["Type"].fillna(str(apps_df["Type"].mode().values[0]),inplace=True)
apps_df["Current Ver"].fillna(str(apps_df["Current Ver"].mode().values[0]),inplace=True)
apps_df["Android Ver"].fillna(str(apps_df["Android Ver"].mode().values[0]),inplace=True)
```

```python
apps_df.isnull().sum()
```

#### Converting "Price", "Review", "Rating" & "Installs" into Numerical value

```python
apps_df["Price"]=apps_df["Price"].apply(lambda x : str(x).replace("$"," ") if "$" in str(x) else str(x))
apps_df["Price"]=apps_df["Price"].apply(lambda x : float(x))
apps_df["Reviews"]=pd.to_numeric(apps_df["Reviews"],errors="coerce")
```

```python
apps_df["Installs"]=apps_df["Installs"].apply(lambda x : str(x).replace("+"," ") if "+" in str(x) else str(x))
apps_df['Installs'] = apps_df['Installs'].str.replace(' ', '').str.replace(',', '')
apps_df["Installs"]=apps_df["Installs"].apply(lambda x : float(x))
```

```python
# grouping the data starting with App and Reviews
apps_df = apps_df.groupby(['App', 'Reviews', 'Category', 'Rating', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', 'Genres', 
                           'Last Updated', 'Current Ver', 'Android Ver'], as_index=False)
```

```python
# reassign Installs values with their mean
apps_df = apps_df['Installs'].mean()
```

```python
# sort the dataframe by Reviews descendingly
apps_df.sort_values(by='Reviews', ascending=False, inplace=True)
```

#### Plotting the graphs of reviews, installs and price per rating

```python

rating_df = apps_df.groupby('Rating').sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].plot(rating_df['Rating'], rating_df['Reviews'], 'r')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Reviews')
axes[0].set_title('Reviews Per Rating')

axes[1].plot(rating_df['Rating'], rating_df['Installs'], 'g')
axes[1].set_xlabel('Rating')
axes[1].set_ylabel('Installs (e+10)')
axes[1].set_title('Installs Per Rating')

axes[2].plot(rating_df['Rating'], rating_df['Price'], 'k')
axes[2].set_xlabel('Rating')
axes[2].set_ylabel('Price')
axes[2].set_title('Price Per Rating')

plt.tight_layout(pad=2)
plt.show()
```

From the above plottings, we can imply that most of the apps with higher rating range of 4.0 - 4.7 are having high amount of reviews and installs. In terms of price, it doesn't reflect a direct relationship with rating, as we could see a fluctuation in term of pricing even at the range of high rating.
